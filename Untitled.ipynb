```python
def try_convert(x):
    try:
        pd.to_numeric(x)
    except:
        if '-' in x:
            try:
                return mean([int(i) for i in x.split('-')])
            except:
                return int(x.split('-')[0])
        elif '+' in x:
            return int(x.split('+')[0])
        else:
            return int(x.split()[0])/12.
```

```python
indivisual['age'] = indivisual['age'].apply(try_convert,)
```
